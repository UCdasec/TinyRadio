```python
import numpy as np
import pandas as pd

# Load the .npy file with allow_pickle=True
file_path = '/home/mabon/UPDATERADIONET/modulation/MODELS_FINAL/Resnet/2/res_out/modelDir/h5_pruning/resnet_Predictions.npy'  # Replace with your file path
data = np.load(file_path, allow_pickle=True).item()  # Use .item() to convert the dictionary to a regular Python dictionary

# Extract predictions and true_labels from the dictionary
predictions = data['predictions']
true_labels = data['true_labels']

# Convert the arrays to pandas DataFrames
df_predictions = pd.DataFrame(predictions)
df_true_labels = pd.DataFrame(true_labels)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import numpy as np
import pandas as pd

# Assuming df_true_labels and df_predictions are DataFrames containing the true labels and predictions respectively

# Convert DataFrame columns to arrays
true_labels = df_true_labels.values.argmax(axis=1)  # Convert one-hot encoded labels to class labels
predictions = df_predictions.values.argmax(axis=1)  # Convert probabilities to class labels

# Create a confusion matrix
conf_matrix = confusion_matrix(true_labels, predictions)

# Define the class labels
class_labels = ['8PSK', 'AM-DSB', 'AM-SSB', 'BPSK', 'CPFSK', 'GFSK', 'PAM4', 'QAM16', 'QAM64', 'QPSK', 'WBFM']

# Convert confusion matrix to a DataFrame for better visualization
conf_matrix_df = pd.DataFrame(conf_matrix, columns=class_labels, index=class_labels)

# Set custom colors for the heatmap
cmap = sns.color_palette("Blues", as_cmap=True)

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix_df, annot=True, fmt="d", cmap=cmap)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

plt.show()
```
